### Analyzing automobile fuel efficiency over time with seaborn

Look at whether there is an overall trend of how miles per gallon (mpg) changes over time on average.

```python
averaged = vehicles.groupby('year')['comb08', 'highway08', 'city08'].agg([np.mean])
```

```python
averaged.columns = ['comb08_mean', 'highway08_mean', 'city08_mean']
```

```python
averaged = averaged.reset_index()
```

```python
averaged.head()
```

```python
averaged.plot(kind = 'scatter', x = 'year', y = 'comb08_mean')
plt.ylabel('Average MPG')
plt.title('All cars');
```

```python
sns.lmplot(x = 'year', y = 'comb08_mean', data = averaged, lowess = True)
plt.ylabel('Average MPG')
plt.title('All cars');
```

This plot might be misleading as hybrid cars with excellent mileage have recently become more popular.

```python
criteria1 = vehicles.fuelType1.isin(['Regular Gasoline', 'Premium Gasoline', 'Midgrade Gasoline'])
criteria2 = vehicles.fuelType2.isnull()
criteria3 = vehicles.atvType != 'Hybrid'
```

```python
vehicles_non_hybrid = vehicles[criteria1 & criteria2 & criteria3]
```

```python
vehicles_non_hybrid.shape
```

```python
averaged_non_hybrid = vehicles_non_hybrid.groupby(['year'])['comb08'].agg([np.mean]).reset_index()
```

```python
averaged_non_hybrid.head()
```

```python
averaged_non_hybrid.plot(kind = 'scatter', x = 'year', y = 'mean')
plt.ylabel('Average MPG')
plt.title('Gasoline cars');
```

There is still a marked rise in the average miles per gallon even after eliminating hybrids.

The next question that we can ask is whether there have been fewer cars with large engines built more recently ?

If this is true, it could explain the increase in average miles per gallon.

Fist, let's verify that larger engine cars have poorer miles per gallon.

```python
vehicles_non_hybrid = vehicles_non_hybrid[vehicles_non_hybrid.displ.notnull()]
```

```python
vehicles_non_hybrid.shape
```

```python
vehicles_non_hybrid.plot(kind = 'scatter', x = 'displ', y = 'comb08')
plt.ylabel('Average MPG')
plt.title('Gasoline cars');
```

```python
avg = vehicles_non_hybrid.groupby('year')['displ', 'comb08'].agg([np.mean]).reset_index()
```

```python
avg.columns = ['year', 'displ_mean', 'comb08_mean']
```

```python
avg.head()
```

```python
sns.pairplot(avg, x_vars = ['year'], y_vars = ['comb08_mean', 'displ_mean'], size = 5);
```
