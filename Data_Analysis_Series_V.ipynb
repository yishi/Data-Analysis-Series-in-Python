```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

# import seaborn could beauty your figures
# becasue seaborn will change some default styles in the matplotlib
import seaborn as sns
```

Read the raw data.

```python
tax_data = pd.read_csv('world_top_income.csv', skiprows = 1)
```

Explore the data.

```python
tax_data.head()
```

```python
tax_data.shape
```

```python
# take a look to determine which countries are involved in our dataset 
tax_data.Country.unique()
```

### Analyzing and visualizing the top income data of the US

```python
tax_us = tax_data[tax_data.Country == 'United States']
```

```python
tax_us.shape
```

```python
tax_us.plot(kind = 'bar', x = 'Year', y = 'Average income per tax unit')
plt.title('U.S. Average Income 1913-2014');
```

```python
tax_us_top = tax_us[['Year', 'Top 10% income share', 'Top 5% income share', 'Top 1% income share', 
        'Top 0.5% income share', 'Top 0.1% income share']]
```

```python
tax_us_top.plot(kind = 'line', x = 'Year')
plt.title('U.S. Percentage Income Share')
plt.ylabel('Percentage');
```

Looking at the raw percentages is useful, but we may also want to consider how the percentages have changed over time, relative to what the average percentage has been for that income group.

```python
tax_us_top.drop(['Year'], axis = 1).mean()
```

```python
tax_us_top_per = tax_us_top.drop(['Year'], axis = 1)/(tax_us_top.drop(['Year'], axis = 1).mean())
```

```python
tax_us_top_per['Year'] = tax_us_top['Year']
```

```python
tax_us_top_per.plot(kind = 'line', x = 'Year')
plt.ylabel('Percentage')
plt.title('Mean Normalized U.S. Percentage Income Share');
```

The dataset also breaks the group's income into categories, such as income that includes capital gains versus income without capital gains.
Let's take a look at how each group's capital gains income fluctuates over time.

```python
tax_us_top_cap = tax_us[['Year', 'Top 10% income share-including capital gains', 
                         'Top 5% income share-including capital gains', 
                         'Top 1% income share-including capital gains', 
                         'Top 0.5% income share-including capital gains', 
                         'Top 0.1% income share-including capital gains']]
```

```python
col_name = ['Year', 'Top 10%', 'Top 5%', 'Top 1%', 'Top 0.5%', 'Top 0.1%']
tax_us_top.columns = col_name
tax_us_top_cap.columns = col_name
```

```python
tax_us_diff = tax_us_top_cap - tax_us_top
```

```python
tax_us_diff['Year'] = tax_us_top['Year']
```

```python
tax_us_diff.plot(kind = 'line', x = 'Year')
plt.ylabel('Percentage Difference')
plt.title('U.S. Capital Gains Income Lift');
```

### Furthering the analysis of the top income groups of the US

```python
tax_us_avg = tax_us[['Year', 'Top 10% average income', 'Top 5% average income', 
                     'Top 1% average income', 'Top 0.5% average income',
                     'Top 0.1% average income']]
```

```python
tax_us_avg.plot(kind = 'line', x = 'Year')
plt.title('U.S. Average Income');
```

We can see how much richer the rich are than the average American:

```python
tax_us_diff = tax_us_avg.drop(['Year'], axis = 1).sub(tax_us_avg['Top 0.1% average income'], axis = 0)
```

```python
tax_us_diff['Year'] = tax_us_avg['Year']
```

```python
tax_us_diff.columns = ['Top 10%', 'Top 5%', 'Top 1%', 'Top 0.5%', 'Top 0.1%', 'Year']
```

```python
tax_us_diff.plot(kind = 'line', x = 'Year')
plt.title('U.S. Income Disparity');
```

In the last analysis, we'll look at the composition of the income of the weathiest Americans. 

```python
tax_us_com = tax_us[['Year', 'Top 10% income composition-Wages, salaries and pensions',
                     'Top 10% income composition-Dividends', 
                     'Top 10% income composition-Interest Income', 
                     'Top 10% income composition-Rents', 
                     'Top 10% income composition-Entrepreneurial income']]
```

```python
tax_us_com.head()
```

```python
tax_us_com.plot(kind = 'area', x = 'Year')
plt.ylabel('Percentage')
plt.title('U.S. Top 10% Income Composition');
```
